```python
# =========================================================
# Amazon Sales Dataset
# 고객 반응 예측 및 최적 할인 전략 추천 시스템
# =========================================================

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

!pip install wordcloud -q
!apt-get -qq install fonts-nanum

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

from IPython.display import display
from wordcloud import WordCloud, STOPWORDS
from PIL import Image, ImageDraw

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# =========================================================
# 0. 디자인 설정
# =========================================================

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fm.fontManager.addfont(font_path)

plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

BG_COLOR = '#F7FAFC'
MAIN_COLOR = '#7FA7D9'
SUB_COLOR = '#BFD7F2'
POINT_COLOR = '#5B8CCB'
HIGHLIGHT_COLOR = '#FF8A8A'
TEXT_COLOR = '#3B4A5A'
GRID_COLOR = '#DCE6F2'

plt.rcParams['figure.facecolor'] = BG_COLOR
plt.rcParams['axes.facecolor'] = BG_COLOR
plt.rcParams['axes.edgecolor'] = '#D0D7E2'
plt.rcParams['grid.color'] = GRID_COLOR


# =========================================================
# 1. 데이터 불러오기
# =========================================================

file_path = '/content/drive/MyDrive/visual/data/amazon.csv'

df = pd.read_csv(file_path)

print('원본 데이터 크기:', df.shape)
display(df.head())


# =========================================================
# 2. 기본 전처리
# =========================================================

def clean_price(series):
    return series.astype(str).replace('[₹,]', '', regex=True)

df['actual_price'] = clean_price(df['actual_price'])
df['discounted_price'] = clean_price(df['discounted_price'])
df['discount_percentage'] = df['discount_percentage'].astype(str).replace('%', '', regex=True)
df['rating_count'] = df['rating_count'].astype(str).replace(',', '', regex=True)

df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

df['category_cleaned'] = df['category'].astype(str).str.split('|').str[0]

df = df.dropna(
    subset=[
        'product_name',
        'actual_price',
        'discount_percentage',
        'rating',
        'rating_count',
        'category_cleaned',
        'about_product',
        'review_title',
        'review_content'
    ]
).copy()

df = df.reset_index(drop=True)

print('전처리 후 데이터 크기:', df.shape)


# =========================================================
# 3. 카테고리 내 price_level 생성
# =========================================================

def get_price_level_by_category(row):
    category = row['category_cleaned']
    price = row['actual_price']

    category_prices = df.loc[df['category_cleaned'] == category, 'actual_price']

    q1 = category_prices.quantile(0.33)
    q2 = category_prices.quantile(0.66)

    if price <= q1:
        return 'Low'
    elif price <= q2:
        return 'Medium'
    else:
        return 'High'

df['price_level'] = df.apply(get_price_level_by_category, axis=1)
df['log_rating_count'] = np.log1p(df['rating_count'])

# 가격 × 할인율 상호작용 항
df['price_discount_interaction'] = df['actual_price'] * df['discount_percentage']


# =========================================================
# 4. 모델 1: 리뷰 수 예측 모델
# =========================================================

X_review = df[
    [
        'actual_price',
        'discount_percentage',
        'price_discount_interaction',
        'category_cleaned',
        'price_level',
        'rating'
    ]
]

y_review = df['log_rating_count']

numeric_review = [
    'actual_price',
    'discount_percentage',
    'price_discount_interaction',
    'rating'
]

categorical_review = [
    'category_cleaned',
    'price_level'
]

numeric_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ]
)

preprocessor_review = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_review),
        ('cat', categorical_transformer, categorical_review)
    ]
)

model_review = Pipeline(
    steps=[
        ('preprocessor', preprocessor_review),
        ('regressor', GradientBoostingRegressor(n_estimators=100, random_state=42))
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X_review,
    y_review,
    test_size=0.2,
    random_state=42
)

model_review.fit(X_train, y_train)

pred_review = model_review.predict(X_test)

rmse_review = np.sqrt(mean_squared_error(y_test, pred_review))
mae_review = mean_absolute_error(y_test, pred_review)
r2_review = r2_score(y_test, pred_review)

print('--- 모델 1 평가 결과 (리뷰 수 예측) ---')
print('RMSE:', round(rmse_review, 4))
print('MAE :', round(mae_review, 4))
print('R²  :', round(r2_review, 4))
print('-----------------------------------------')


# =========================================================
# 5. 모델 2: 평점 예측 모델
# =========================================================

X_rating = df[
    [
        'actual_price',
        'discount_percentage',
        'price_discount_interaction',
        'category_cleaned',
        'price_level'
    ]
]

y_rating = df['rating']

numeric_rating = [
    'actual_price',
    'discount_percentage',
    'price_discount_interaction'
]

categorical_rating = [
    'category_cleaned',
    'price_level'
]

preprocessor_rating = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_rating),
        ('cat', categorical_transformer, categorical_rating)
    ]
)

model_rating = Pipeline(
    steps=[
        ('preprocessor', preprocessor_rating),
        ('regressor', GradientBoostingRegressor(n_estimators=100, random_state=42))
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X_rating,
    y_rating,
    test_size=0.2,
    random_state=42
)

model_rating.fit(X_train, y_train)

pred_rating = model_rating.predict(X_test)

rmse_rating = np.sqrt(mean_squared_error(y_test, pred_rating))
mae_rating = mean_absolute_error(y_test, pred_rating)
r2_rating = r2_score(y_test, pred_rating)

print('--- 모델 2 평가 결과 (평점 예측) ---')
print('RMSE:', round(rmse_rating, 4))
print('MAE :', round(mae_rating, 4))
print('R²  :', round(r2_rating, 4))
print('-----------------------------------------')


# =========================================================
# 6. 리뷰 분석 설정
# =========================================================
# 개선된 ABSA 설정
# =========================================================

aspect_keywords = {
    '가격/가성비': ['price', 'cheap', 'expensive', 'value', 'worth', 'budget', 'cost'],
    '품질': ['quality', 'durable', 'material'],
    '배송/포장': ['delivery', 'shipping', 'package', 'packing', 'arrived'],
    '사용성': ['easy', 'comfortable', 'convenient', 'install'],
    '디자인': ['design', 'look', 'style', 'color'],
    '성능': ['performance', 'fast', 'battery', 'sound', 'noise', 'power']
}

positive_words = [
    'good', 'great', 'excellent', 'amazing',
    'fast', 'love', 'perfect', 'easy',
    'comfortable', 'nice', 'best', 'awesome',
    'satisfied', 'worth'
]

negative_words = [
    'bad', 'poor', 'terrible', 'worst',
    'slow', 'hate', 'broken', 'problem',
    'difficult', 'hard', 'noise', 'damaged',
    'waste', 'issue'
]


# =========================================================
# 개선된 ABSA 함수
# =========================================================

def extract_aspect_sentiment(similar_products):
    rows = []

    for _, row in similar_products.iterrows():
        text = (
            str(row['review_title']) + ' ' +
            str(row['review_content'])
        ).lower()

        words = text.split()

        for aspect, keywords in aspect_keywords.items():
            positive_count = 0
            negative_count = 0

            for keyword in keywords:
                for i, word in enumerate(words):
                    if keyword in word:
                        # 키워드 주변 윈도우 생성 (-3 ~ +3 단어) 탐색으로 오탐지 차단
                        start = max(0, i - 3)
                        end = min(len(words), i + 4)
                        nearby_words = words[start:end]

                        for nearby_word in nearby_words:
                            if nearby_word in positive_words:
                                positive_count += 1
                            elif nearby_word in negative_words:
                                negative_count += 1

            total_count = positive_count + negative_count

            if total_count > 0:
                if positive_count > negative_count:
                    sentiment = '긍정'
                elif negative_count > positive_count:
                    sentiment = '부정'
                else:
                    sentiment = '중립'

                rows.append([
                    aspect,
                    sentiment,
                    total_count
                ])

    if len(rows) == 0:
        return pd.DataFrame(columns=['속성', '감정', '언급 횟수'])

    return pd.DataFrame(rows, columns=['속성', '감정', '언급 횟수'])


# =========================================================
# 7. 검색 키워드 예시 출력 함수
# =========================================================

def show_search_examples():
    print('\n==============================')
    print('검색 키워드 예시')
    print('==============================')

    example_keywords = [
        'earbuds', 'headphones', 'speaker', 'soundbar',
        'watch', 'smartwatch', 'mouse', 'keyboard',
        'laptop', 'monitor', 'charger', 'cable',
        'phone', 'camera', 'tv', 'remote',
        'dryer', 'fan', 'mixer', 'coffee',
        'kettle', 'iron', 'lamp', 'bulb',
        'bag', 'bottle', 'stand', 'cover'
    ]

    print(', '.join(example_keywords))

    print('\n카테고리 예시')
    category_table = (
        df['category_cleaned']
        .value_counts()
        .reset_index()
    )

    category_table.columns = ['카테고리', '상품 수']
    display(category_table.head(10))

    print('\n실제 상품명 예시')
    sample_products = (
        df[
            [
                'product_name',
                'category_cleaned',
                'actual_price',
                'rating'
            ]
        ]
        .sample(10, random_state=42)
        .reset_index(drop=True)
    )

    sample_products.columns = [
        '상품명',
        '카테고리',
        '가격',
        '평점'
    ]

    display(sample_products)


# =========================================================
# 8. 상품 검색 함수
# =========================================================

def search_products():
    show_search_examples()

    keyword = input('\n검색할 상품 키워드를 입력하세요: ').strip().lower()

    result = df[
        df['product_name'].str.lower().str.contains(keyword, na=False)
    ].copy()

    if len(result) < 5:
        result = df[
            df['product_name'].str.lower().str.contains(keyword, na=False)
            | df['about_product'].str.lower().str.contains(keyword, na=False)
            | df['category_cleaned'].str.lower().str.contains(keyword, na=False)
        ].copy()

    if len(result) == 0:
        print('검색 결과가 없습니다. 다른 키워드로 다시 검색해주세요.')
        return None

    result = result.head(20).copy()
    result = result.reset_index(drop=True)
    result['선택번호'] = range(len(result))

    display_table = result[
        [
            '선택번호',
            'product_name',
            'category_cleaned',
            'actual_price',
            'discount_percentage',
            'rating',
            'rating_count'
        ]
    ].copy()

    display_table.columns = [
        '선택번호',
        '상품명',
        '카테고리',
        '가격',
        '현재 할인율',
        '평점',
        '리뷰 수'
    ]

    display(display_table)

    selected_num = int(input('분석할 상품의 선택번호를 입력하세요: '))
    selected_product = result[result['선택번호'] == selected_num].iloc[0]

    return selected_product


# =========================================================
# 9. 선택 상품 분석 함수
# =========================================================

def analyze_selected_product(selected_product):
    input_category = selected_product['category_cleaned']
    input_price = float(selected_product['actual_price'])
    input_about_product = selected_product['about_product']
    input_product_name = selected_product['product_name']

    print('\n선택한 상품 정보')
    print('상품명:', input_product_name)
    print('카테고리:', input_category)
    print('가격:', f'{int(input_price):,}')
    print('상품 설명:', input_about_product[:300], '...')

    discount_options = list(range(0, 51, 5))

    category_prices = df.loc[df['category_cleaned'] == input_category, 'actual_price']
    q1 = category_prices.quantile(0.33)
    q2 = category_prices.quantile(0.66)

    if input_price <= q1:
        input_price_level = 'Low'
    elif input_price <= q2:
        input_price_level = 'Medium'
    else:
        input_price_level = 'High'

    results = []

    for discount in discount_options:
        discounted_price = input_price * (1 - discount / 100)
        interaction = input_price * discount

        input_data_rating = pd.DataFrame({
            'actual_price': [input_price],
            'discount_percentage': [discount],
            'price_discount_interaction': [interaction],
            'category_cleaned': [input_category],
            'price_level': [input_price_level]
        })

        predicted_rating = model_rating.predict(input_data_rating)[0]
        predicted_rating = min(max(predicted_rating, 1), 5)

        input_data_review = pd.DataFrame({
            'actual_price': [input_price],
            'discount_percentage': [discount],
            'price_discount_interaction': [interaction],
            'category_cleaned': [input_category],
            'price_level': [input_price_level],
            'rating': [predicted_rating]
        })

        predicted_log_review = model_review.predict(input_data_review)[0]
        predicted_review_count = max(0, int(np.expm1(predicted_log_review)))

        results.append([
            discount,
            int(discounted_price),
            round(predicted_rating, 2),
            predicted_review_count
        ])

    result_df = pd.DataFrame(
        results,
        columns=[
            '할인율(%)',
            '할인 후 가격',
            '예상 평점',
            '예상 리뷰 수'
        ]
    )

    result_df['리뷰수_정규화'] = result_df['예상 리뷰 수'] / result_df['예상 리뷰 수'].max()
    result_df['평점_정규화'] = result_df['예상 평점'] / 5

    result_df['리뷰증가율'] = (
        result_df['예상 리뷰 수'] - result_df['예상 리뷰 수'].min()
    ) / (
        result_df['예상 리뷰 수'].max() - result_df['예상 리뷰 수'].min() + 1e-9
    )

    result_df['예상매출지표'] = result_df['할인 후 가격'] * result_df['예상 리뷰 수']
    result_df['매출_정규화'] = result_df['예상매출지표'] / result_df['예상매출지표'].max()

    result_df['추천점수'] = (
        result_df['평점_정규화'] * 0.3
        + result_df['리뷰수_정규화'] * 0.3
        + result_df['매출_정규화'] * 0.4
    )

    best = result_df.sort_values(by='추천점수', ascending=False).iloc[0]

    # 유사 상품 찾기
    df['similarity_text'] = (
        df['category_cleaned'].astype(str)
        + ' '
        + df['about_product'].astype(str)
    )

    input_text = input_category + ' ' + input_about_product

    vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = vectorizer.fit_transform(df['similarity_text'])
    input_vector = vectorizer.transform([input_text])

    similarity_scores = cosine_similarity(input_vector, tfidf_matrix).flatten()
    df['similarity_score'] = similarity_scores

    similar_products = (
        df[df['product_name'] != input_product_name]
        .sort_values(by='similarity_score', ascending=False)
        .head(10)
        .copy()
    )

    similar_review_text = (
        similar_products['review_title'].astype(str)
        + ' '
        + similar_products['review_content'].astype(str)
    )

    all_review_text = ' '.join(similar_review_text).lower()

    keyword_rows = []
    for aspect, keywords in aspect_keywords.items():
        count = 0
        for word in keywords:
            count += all_review_text.count(word)
        keyword_rows.append([aspect, count])

    keyword_df = pd.DataFrame(keyword_rows, columns=['키워드 유형', '언급 횟수'])
    keyword_df = keyword_df.sort_values(by='언급 횟수', ascending=False)

    absa_raw = extract_aspect_sentiment(similar_products)

    if len(absa_raw) > 0:
        absa_summary = (
            absa_raw
            .groupby(['속성', '감정'])['언급 횟수']
            .sum()
            .reset_index()
        )
    else:
        absa_summary = pd.DataFrame(columns=['속성', '감정', '언급 횟수'])

    # 최종 출력
    print('\n===== 할인율별 예측 결과 =====')
    display(result_df)

    print('\n===== 최종 추천 결과 =====')
    print(f"추천 할인율: {best['할인율(%)']}% ")
    print(f"할인 후 가격: {best['할인 후 가격']:,.0f}")
    print(f"예상 평점: {best['예상 평점']}")
    print(f"예상 리뷰 수: {best['예상 리뷰 수']:,}개")

    print('\n해석:')
    print(f"이 상품은 {best['할인율(%)']}% 할인에서 예상 평점, 리뷰 수 및 판매자 수익성 지표를 종합한 추천점수가 가장 높습니다.")

    print('\n===== 유사 상품 TOP 10 =====')
    display(
        similar_products[
            [
                'product_name',
                'category_cleaned',
                'actual_price',
                'discount_percentage',
                'rating',
                'rating_count',
                'similarity_score'
            ]
        ].reset_index(drop=True)
    )

    print('\n===== 유사 상품 리뷰 주요 속성 =====')
    display(keyword_df.reset_index(drop=True))

    print('\n===== ABSA 기반 속성별 리뷰 반응 =====')
    display(absa_summary.reset_index(drop=True))

    # =====================================================
    # 대시보드 시각화
    # =====================================================

    # (1) 할인율별 예상 리뷰 수
    plt.figure(figsize=(8, 5))
    plt.plot(
        result_df['할인율(%)'],
        result_df['예상 리뷰 수'],
        marker='o',
        linewidth=3,
        color=MAIN_COLOR,
        markersize=9
    )

    plt.axvline(
        best['할인율(%)'],
        linestyle='--',
        color=HIGHLIGHT_COLOR,
        linewidth=2,
        label='추천 할인율'
    )

    for _, row in result_df.iterrows():
        plt.text(
            row['할인율(%)'],
            row['예상 리뷰 수'],
            f"{row['예상 리뷰 수']:,}",
            ha='center',
            va='bottom',
            color=TEXT_COLOR
        )

    plt.title('할인율별 예상 리뷰 수', fontsize=15, fontweight='bold', color=TEXT_COLOR)
    plt.xlabel('할인율(%)')
    plt.ylabel('예상 리뷰 수')
    plt.grid(True)
    plt.legend()
    plt.show()

    # (2) 할인율별 예상 평점
    plt.figure(figsize=(8, 5))
    plt.plot(
        result_df['할인율(%)'],
        result_df['예상 평점'],
        marker='o',
        linewidth=3,
        color=POINT_COLOR,
        markersize=9
    )

    plt.axvline(
        best['할인율(%)'],
        linestyle='--',
        color=HIGHLIGHT_COLOR,
        linewidth=2,
        label='추천 할인율'
    )

    for _, row in result_df.iterrows():
        plt.text(
            row['할인율(%)'],
            row['예상 평점'],
            f"{row['예상 평점']:.2f}",
            ha='center',
            va='bottom',
            color=TEXT_COLOR
        )


    plt.title('할인율별 예상 평점', fontsize=15, fontweight='bold', color=TEXT_COLOR)
    plt.xlabel('할인율(%)')
    plt.ylabel('예상 평점')
    plt.grid(True)
    plt.legend()
    plt.show()

    # (3) 시장 내 상품 위치
    plt.figure(figsize=(9, 6))
    scatter = plt.scatter(
        df['actual_price'],
        df['log_rating_count'],
        c=df['rating'],
        cmap='Blues',
        alpha=0.55,
        s=55
    )
    plt.colorbar(scatter, label='평점')

    plt.scatter(
        input_price,
        np.log1p(best['예상 리뷰 수']),
        s=450,
        marker='*',
        color=HIGHLIGHT_COLOR,
        edgecolor='white',
        linewidth=2,
        label='선택 상품 예상 위치'
    )

    plt.title('시장 내 상품 위치', fontsize=15, fontweight='bold', color=TEXT_COLOR)
    plt.xlabel('가격')
    plt.ylabel('log(리뷰 수 + 1)')
    plt.grid(True)
    plt.legend()
    plt.show()

    # (4-1) 유사 상품 리뷰 주요 속성 Bar Chart
    keyword_plot_df = keyword_df.sort_values(by='언급 횟수', ascending=True).reset_index(drop=True)

    plt.figure(figsize=(8, 5))
    plt.barh(
        keyword_plot_df['키워드 유형'],
        keyword_plot_df['언급 횟수'],
        color=SUB_COLOR
    )

    for i, value in enumerate(keyword_plot_df['언급 횟수']):
        plt.text(value, i, str(value), va='center', color=TEXT_COLOR)

    plt.title('유사 상품 리뷰 주요 속성', fontsize=15, fontweight='bold', color=TEXT_COLOR)
    plt.xlabel('언급 횟수')
    plt.ylabel('키워드 유형')
    plt.grid(axis='x')
    plt.show()

    # (4-2) 쇼핑카트 실루엣 WordCloud
    custom_stopwords = set(STOPWORDS)
    custom_stopwords.update([
        'product', 'one', 'will', 'also', 'use', 'using',
        'amazon', 'really', 'got', 'get', 'much', 'even',
        'good', 'nice'
    ])

    if len(all_review_text.strip()) > 0:
        width, height = 1000, 650
        mask_img = Image.new('L', (width, height), 255)
        draw = ImageDraw.Draw(mask_img)

        draw.polygon([(220, 180), (820, 180), (730, 470), (300, 470)], fill=0)
        draw.line([(160, 120), (240, 180)], fill=0, width=40)
        draw.line([(320, 500), (710, 500)], fill=0, width=28)
        draw.ellipse((280, 500, 400, 620), fill=0)
        draw.ellipse((620, 500, 740, 620), fill=0)

        cart_mask = np.array(mask_img)

        wordcloud = WordCloud(
            font_path=font_path,
            width=width,
            height=height,
            background_color=BG_COLOR,
            colormap='Blues',
            max_words=120,
            stopwords=custom_stopwords,
            mask=cart_mask,
            contour_width=3,
            contour_color='#7FA7D9',
            prefer_horizontal=0.9
        ).generate(all_review_text)

        plt.figure(figsize=(11, 7))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('유사 상품 리뷰 WordCloud', fontsize=18, fontweight='bold', color=TEXT_COLOR)
        plt.show()
    else:
        print('WordCloud를 생성할 리뷰 텍스트가 충분하지 않습니다.')

    # (4-3) ABSA 속성별 감정 분석
    if len(absa_summary) > 0:
        pivot_absa = absa_summary.pivot_table(
            index='속성',
            columns='감정',
            values='언급 횟수',
            aggfunc='sum',
            fill_value=0
        )

        for col in ['긍정', '중립', '부정']:
            if col not in pivot_absa.columns:
                pivot_absa[col] = 0

        pivot_absa = pivot_absa[['긍정', '중립', '부정']]

        pivot_absa.plot(
            kind='bar',
            stacked=True,
            figsize=(9, 5),
            color=['#7FA7D9', '#C8D6E5', '#FF8A8A']
        )

        plt.title('ABSA 기반 속성별 리뷰 반응', fontsize=15, fontweight='bold', color=TEXT_COLOR)
        plt.xlabel('속성')
        plt.ylabel('언급 횟수')
        plt.xticks(rotation=0)
        plt.grid(axis='y')
        plt.legend(title='감정')
        plt.show()
    else:
        print('ABSA 분석 결과가 충분하지 않습니다.')

    # (5) 추천 할인율 강조 Bar Chart
    plt.figure(figsize=(8, 5))

    colors = [
        HIGHLIGHT_COLOR if x == best['할인율(%)'] else SUB_COLOR
        for x in result_df['할인율(%)']
    ]

    plt.bar(
        result_df['할인율(%)'].astype(str),
        result_df['추천점수'],
        color=colors
    )

    for i, row in result_df.iterrows():
        plt.text(
            i,
            row['추천점수'],
            f"{row['추천점수']:.3f}",
            ha='center',
            va='bottom',
            color=TEXT_COLOR
        )

    plt.title('추천 할인율 강조', fontsize=15, fontweight='bold', color=TEXT_COLOR)
    plt.xlabel('할인율(%)')
    plt.ylabel('추천 점수')
    plt.grid(axis='y')
    plt.show()

    # =====================================================
    # 판매자 전략 코멘트
    # =====================================================

    top_aspect = keyword_df.sort_values(by='언급 횟수', ascending=False).iloc[0]['키워드 유형']

    if best['할인율(%)'] <= 10:
        discount_title = f"{best['할인율(%)']}% 할인 전략 추천"
        discount_comment = (
            "낮은 할인율에서도 예상 평점과 리뷰 수가 안정적으로 유지되는 것으로 나타났습니다. "
            "따라서 과도한 할인보다는 수익성을 높은 수준으로 유지하는 방향의 마진 방어 전략이 적합합니다."
        )
    elif best['할인율(%)'] <= 20:
        discount_title = f"{best['할인율(%)']}% 균형형 할인 전략 추천"
        discount_comment = (
            "중간 수준의 할인율에서 예상 리뷰 수와 평점이 균형 있게 유지되는 것으로 나타났습니다. "
            "가격 경쟁력과 마진 확보를 동시에 가져가는 균형 잡힌 밸런싱 판매 전략이 요구됩니다."
        )
    else:
        discount_title = f"{best['할인율(%)']}% 프로모션형 할인 전략 추천"
        discount_comment = (
            "높은 할인율에서 예상 리뷰 수 증가 효과가 크게 나타났습니다. "
            "단기 노출 확대, 박리다매 마케팅 및 리뷰 대량 확보 전략 관점에 적합합니다."
        )

    aspect_title = f"{top_aspect} 강조 전략"
    aspect_comments = {
        '가격/가성비': "유사 상품 리뷰에서 가격 대비 만족도 관련 언급이 많았습니다. 가성비와 할인 혜택을 강조하는 것이 효과적일 수 있습니다.",
        '품질': "유사 상품 리뷰에서 품질 관련 언급이 많았습니다. 내구성, 안정성, 제품 완성도를 강조하는 것이 좋습니다.",
        '배송/포장': "배송 속도와 포장 상태 관련 언급이 많았습니다. 빠른 배송 및 안전한 포장 경험을 강조하는 것이 좋습니다.",
        '사용성': "유사 상품 리뷰에서 편의성과 조작 편리성 언급이 많았습니다. 간편한 실사용 경험을 소구하는 것이 효과적입니다.",
        '디자인': "디자인과 외형 관련 언급이 많았습니다. 제품의 독창적인 스타일과 시각적 요소를 강조하는 것이 좋습니다.",
        '성능': "유사 상품 리뷰에서 성능 관련 언급이 가장 많이 나타났습니다. 속도, 배터리 효율 등 핵심 스펙을 전면에 내세우는 전략이 주효합니다."
    }
    aspect_comment = aspect_comments.get(top_aspect, "유사 상품 리뷰에서 해당 속성 관련 언급이 많았습니다.")

    print('\n========================================')
    print('판매자 전략 분석 결과')
    print('========================================')
    print(f'\n[ {discount_title} ]')
    print(discount_comment)
    print(f'\n[ {aspect_title} ]')
    print(aspect_comment)
    print('\n추가 인사이트')
    print(f"현재 분석 결과 기준으로 예상 평점은 {best['예상 평점']}점, 예상 리뷰 수는 약 {best['예상 리뷰 수']:,}개로 예측됩니다.")

    # =====================================================
    # 최종 판매 전략 리포트 생성기
    # =====================================================

    price_percentile = (df['actual_price'] < input_price).mean() * 100
    review_percentile = (df['rating_count'] < best['예상 리뷰 수']).mean() * 100
    rating_percentile = (df['rating'] < best['예상 평점']).mean() * 100

    # 전략 유형 결정
    if best['할인율(%)'] <= 10:
        strategy_type = '수익성 유지형 전략'
    elif best['할인율(%)'] <= 20:
        strategy_type = '균형형 전략'
    else:
        strategy_type = '공격형 프로모션 전략'

    # 속성별 전략 코멘트
    report_aspect_comments = {
        '가격/가성비': '소비자들은 가격 대비 만족도를 중요하게 평가하는 경향이 나타났습니다. 가격 경쟁력과 할인 혜택을 부각하는 전략이 필수적입니다.',
        '품질': '리뷰에서는 제품 품질과 내구성 관련 언급 비중이 높게 나타났습니다. 상세페이지에서 품질 신뢰성 검증 자료를 전면에 내세우는 것이 중요합니다.',
        '배송/포장': '배송 속도와 포장 상태 관련 반응이 주로 파악되었습니다. 신속하고 꼼꼼한 물류 인프라 장점을 마케팅 포인트로 삼으십시오.',
        '사용성': '소비자들은 사용 편의성과 직관적인 작동법을 귀중하게 인식합니다. 조작 가이드나 사용 튜토리얼 영상 중심의 마케팅이 유효합니다.',
        '디자인': '디자인과 세련된 외형 관련 핵심 피드백이 많았습니다. 감성적인 비주얼 컷과 고품질의 제품 연출 이미지를 전면에 배치하십시오.',
        '성능': '유사 상품 리뷰에서는 핵심 스펙 및 성능 피드백이 가장 높게 나타났습니다. 배터리 타임, 출력 속도 등의 하드웨어 성능을 기술적으로 소구하는 전략이 유리합니다.'
    }
    report_aspect_comment = report_aspect_comments.get(top_aspect, '유사 상품 리뷰에서 해당 속성 피드백이 활발히 나타났습니다.')

    # 포스터형 리포트 생성
    fig = plt.figure(figsize=(20, 11))
    fig.patch.set_facecolor(BG_COLOR)

    plt.suptitle('AI 기반 Amazon 상품 판매 전략 리포트', fontsize=30, fontweight='bold', color=TEXT_COLOR, y=0.98)
    plt.figtext(0.5, 0.935, f"분석 상품 : {input_product_name[:80]}", ha='center', fontsize=15, color='#5B6B7A')

    # 1. 할인율별 리뷰수
    ax1 = plt.subplot2grid((3, 4), (0, 0), colspan=2)
    ax1.plot(result_df['할인율(%)'], result_df['예상 리뷰 수'], marker='o', linewidth=3, color=MAIN_COLOR)
    ax1.axvline(best['할인율(%)'], linestyle='--', color=HIGHLIGHT_COLOR, linewidth=2)
    ax1.set_title('할인율별 예상 리뷰 수', fontsize=16, fontweight='bold')
    ax1.set_xlabel('할인율(%)')
    ax1.set_ylabel('예상 리뷰 수')
    ax1.grid(True)

    # 2. 할인율별 평점
    ax2 = plt.subplot2grid((3, 4), (0, 2), colspan=2)
    ax2.plot(result_df['할인율(%)'], result_df['예상 평점'], marker='o', linewidth=3, color=POINT_COLOR)
    ax2.axvline(best['할인율(%)'], linestyle='--', color=HIGHLIGHT_COLOR, linewidth=2)
    ax2.set_title('할인율별 예상 평점', fontsize=16, fontweight='bold')
    ax2.set_xlabel('할인율(%)')
    ax2.set_ylabel('예상 평점')
    ax2.grid(True)

    # 3. 시장 내 위치
    ax3 = plt.subplot2grid((3, 4), (1, 0), colspan=2)
    ax3.scatter(df['actual_price'], np.log1p(df['rating_count']), c=df['rating'], cmap='Blues', alpha=0.5, s=40)
    ax3.scatter(input_price, np.log1p(best['예상 리뷰 수']), s=500, marker='*', color=HIGHLIGHT_COLOR, edgecolor='white', linewidth=2)
    ax3.set_title('시장 내 상품 위치', fontsize=16, fontweight='bold')
    ax3.set_xlabel('가격')
    ax3.set_ylabel('log(리뷰 수 + 1)')
    ax3.grid(True)

    # 4. 주요 리뷰 속성
    ax4 = plt.subplot2grid((3, 4), (1, 2), colspan=2)
    ax4.barh(keyword_plot_df['키워드 유형'], keyword_plot_df['언급 횟수'], color=SUB_COLOR)
    ax4.set_title('유사 상품 리뷰 주요 속성', fontsize=16, fontweight='bold')

    # 5. 판매 전략 텍스트
    ax5 = plt.subplot2grid((3, 4), (2, 0), colspan=4)
    ax5.axis('off')

    strategy_text = f"""
[ 추천 할인 전략 ]

추천 할인율 : {best['할인율(%)']}%
전략 유형 : {strategy_type}

현재 상품은 해당 할인율을 적용했을 때, 예상 평점과 유입 리뷰 규모, 그리고 매출 볼륨의 밸런스가 최고점에 도달했습니다.


[ 시장 내 위치 분석 ]

가격 상위 : 약 {price_percentile:.1f}% | 리뷰 수 상위 : 약 {review_percentile:.1f}% | 평점 상위 : 약 {rating_percentile:.1f}%


[ 소비자 반응 분석 ]

유사 대조군 상품 리뷰 검토 결과, '{top_aspect}' 관련 소비자 소통 강도가 제일 강력하게 집계되었습니다.
{report_aspect_comment}


[ 최종 마케팅 오퍼레이션 ]

예상 평점 편차 : {best['예상 평점']}점 | 예상 리뷰 확보 수치 : 약 {best['예상 리뷰 수']:,}개
정량적 AI 시뮬레이션 결과에 기반하여, 본 상품은 마진 방어선에 걸맞은 추천 할인율을 엄수함과 동시에 타겟 마케팅 시 '{top_aspect}' 핵심 키워드를 최우선 셀링포인트로 강조할 것을 강력히 권고합니다.
"""

    ax5.text(
        0.01, 0.95, strategy_text, fontsize=14, color=TEXT_COLOR, verticalalignment='top',
        bbox=dict(facecolor='white', edgecolor='#DCE6F2', boxstyle='round,pad=1')
    )

    plt.tight_layout(rect=[0, 0, 1, 0.95])

    # 이미지 저장
    save_name = input_product_name[:30].replace('/', '_').replace(':', '_').strip()
    fig.savefig(f'{save_name}_final_strategy_report.png', dpi=300, bbox_inches='tight', facecolor=BG_COLOR)

    print('\n최종 판매 전략 리포트 저장 완료!')
    print(f'{save_name}_final_strategy_report.png')

    plt.show()


# =========================================================
# 10. 여러 상품 반복 분석 실행
# =========================================================

while True:
    selected_product = search_products()

    if selected_product is not None:
        analyze_selected_product(selected_product)

    again = input('\n다른 상품도 분석할까요? y/n : ').strip().lower()
    if again != 'y':
        print('분석을 종료합니다.')
        break
```
